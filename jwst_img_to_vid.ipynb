```python
from bs4 import BeautifulSoup
import requests
import os
from PIL import Image
import json
```

```python
def resize_image(image_path, target_width, target_height, background_color=(0, 0, 0, 0)):
    # Open the image
    image = Image.open(image_path)

    # Calculate the aspect ratio
    original_width, original_height = image.size
    aspect_ratio = original_width / original_height

    # Determine new size while maintaining aspect ratio
    if original_width > original_height:
        new_width = min(target_width, original_width)
        new_height = int(new_width / aspect_ratio)
    else:
        new_height = min(target_height, original_height)
        new_width = int(new_height * aspect_ratio)

    # Resize the image
    resized_image = image.resize((new_width, new_height), Image.ANTIALIAS)

    # Create a new image with the target size and background color
    new_image = Image.new('RGB', (target_width, target_height), background_color)

    # Calculate position to paste the resized image
    x = (target_width - new_width) // 2
    y = (target_height - new_height) // 2

    # Paste the resized image onto the new image
    new_image.paste(resized_image, (x, y))

    return new_image

def download_file(url, filename):
    '''
    Downloads the file to the current directory
    '''
    try:
        # Send a GET request to the URL
        response = requests.get(url, stream=True)

        # Raise an exception if the request was unsuccessful
        response.raise_for_status()

        # Extract the file extension from the URL
        _, extension = os.path.splitext(url)

        # If the provided filename doesn't have an extension, append the original one
        if not os.path.splitext(filename)[1]:
            filename += extension

        # Save the file
        with open(filename, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)

        return f"File saved as {filename}"

    except requests.exceptions.RequestException as e:
        return f"Error: {e}"

def url_parse(url):
    '''
    Given the URL, this function downloads the observations and associated
    metadata
    '''
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # this is where we find the download links for the jswt observations
        media_library = soup.find('div', class_='media-library-links-list')
        print(media_library)

        # the index name will be the title from the html file
        title_raw = soup.find('title').get_text(strip=True)
        title = title_raw.replace('/', 'Or').replace(' ', '_')
        filename = f'[python]{title}'

        if media_library:
            observations = media_library.find_all('a')
            print(observations)
            if observations:
                target_image = 'http:' + observations[-1].get('href')
                # Save the content
                metadata = {
                    'html': response.text,
                    'filename': filename
                }
                # write the json file
                with open(f"{filename}.json", 'w') as file:
                    json.dump(metadata, file)
                # write the png file
                download_file(target_image, filename)

                return title
            else:
                print("No links found in the specified div.")
        else:
            print("Div with class 'media-library-links-list' not found.")
    else:
        print(f"Failed to retrieve content. Status code: {response.status_code}")

def scrape():
    urls = links
    for link in urls:
        result = url_parse(link)
        print(f'Done with {result}')

def resize():
    # define dimensions
    h = 576
    w = 1024
    # go through current directory
    for file in os.listdir():
        # check to see if this file is an image
        if file.split('.')[-1] in ['png', 'jpg', 'tif'] :
            print(file)
            resized = resize_image(file, w, h)
            rename = f'[resized]{file}'
            resized.save(rename)
```
